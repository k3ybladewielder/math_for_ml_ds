# System of Linear Equations

## System of Sentences

* Um sistema de sentenças é análogo a um sistema de equações, já que ele traz informações.
* Um sistema linear é composto por duas ou mais equações lineares com variáveis. Um sistema linear é dito não singular se ele tem uma única solução única, ou seja, se as equações do sistema formam uma interseção única. Nesse caso, as equações são independentes e não formam uma reta paralela.
* Já um sistema linear singular é aquele em que as equações do sistema formam uma reta paralela, ou seja, não há interseção entre elas. Isso significa que não há uma solução única para o sistema, e portanto, não há valores únicos para as variáveis.
* Como exemplo, assuma três pares de frases como:
* Sistema 1:
<center>"O cachorro é preto"</center>
<center>"O gato é laranja"</center>

Esse sistema **completo** e **não singular**.

* Sistema 2:
<center>"O cachorro é preto"</center>
<center>"O cachorro é preto"</center>

Esse sistema **redundante** e **singular**.

<center>"O cachorro é preto"</center>
<center>"O cachorro é branco"</center>

Esse sistema **contraditório** e **singular**.

* Um sistema completo é chamado de não-singular. 
* Um sistema redundante ou contraditório é chamado de singular
* Um sistema não singular é aquele que carrega o maior número de informações, enquanto que os singulares possem menos e são menos informativos.
* Um sistema de sentenças pode possuir várias sentenças
* A redundância de informação pode ser medida com o rank

## System of equations

* Linear algebra é o estudo de equações lineares.
* Em equações lineares somente podemos ter número/escalares junto das variáveis (ex. 2a + 3b = 15)
* Enquanto que equações não lineares são mais complexas, podendo ter potência (ex. $a^{2} + b^{2} = 10$), seno (ex. $sin(a) + b^{5} = 15$), tangente, etc.  
* Um sistema de equações lineares é um conjunto finito de equações lineares aplicadas num mesmo conjunto, igualmente finito, de variáveis.
* Considere os seguintes sitemas:

**Sistema 1:**
<center>$$ a + b = 10 $$</center>
<center>$$ a + 2b = 12 $$</center>
A única solução é a = 8 e b = 2. O sistema é completo e não-singular

**Sistema 2:**
<center>$$ a + b = 10 $$</center>
<center>$$ 2a + 2b = 20 $$</center>
Existem infinitas soluções, uma para cada variação de valor que resulte em 20. E o sistema é redundante e singular.

**Sistema 3:**
<center>$$ a + b = 10 $$</center>
<center>$$ 2a + 2b = 24 $$</center>
Esse sistema não possui solução porquê as duas equações se contradizem. E o sistema é contraditório e singular

## System of equations as lines

* Equações lineares podem ser representadas por linhas num plano carteziano quanto temos 2 variáveis, com 3 ou mais teremos planos (hiperplanos) num espaço.
* Sistemas de equações lineares podem ser representadas como conjunto de linhas num plano. Dessa forma, conseguimos visualizar as soluções e quando são singulares ou não singulares
* Na seguinte equação $ a + b = 10 $, podemos passar a reta pelos seguintes pontos: $ (-4, 14), (0, 10), (4, 6), (8, 2), (10, 0), (12, -2) $. Cada ponto é uma solução para essa equação.

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 6.0)

x = [-4, 0, 4, 8, 10, 12]
y = [14, 10, 6, 2, 0, -2]

fig, ax = plt.subplots()
ax.plot(x, y, '-o')
ax.grid(linestyle='--')
ax.axhline(0, color='black', lw=1)
ax.axvline(0, color='black', lw=1)
ax.title.set_text("Equação 1, Sistema 1")
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
plt.show()
```

* Já na equação $ a + 2b = 12$, podemos passar a reta pelos seguintes pontos: $(0, 6), (12, 0), (8, 2), (-4, 8)$

```python
import matplotlib.pyplot as plt

x = [0, 12, 8, -4]
y = [6, 0, 2, 8]

fig, ax = plt.subplots()
ax.plot(x, y, '-o')
ax.grid(linestyle='--')
ax.axhline(0, color='black', lw=1)
ax.axvline(0, color='black', lw=1)
ax.title.set_text("Equação 2, Sistema 1")
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
plt.show()
```

* Um sistema de quações com 2 equações possue duas retas no mesmo plano. Utilizando as 4 coordenadas dos plots acima, a única solução para esse sistema com essas 2 equações é $(8, 2)$, ou seja, o ponto em que as retas se cruzam.

```python
# import matplotlib.pyplot as plt

x1 = [-4, 0, 4, 8, 10, 12]
y1 = [14, 10, 6, 2, 0, -2]
x2 = [0, 12, 8, -4]
y2 = [6, 0, 2, 8]

fig, ax = plt.subplots()
ax.plot(x1, y1, '-o', label='Série 1')
ax.plot(x2, y2, '-s', label='Série 2')
ax.grid(linestyle='--')
ax.axhline(0, color='black', lw=1)
ax.axvline(0, color='black', lw=1)
ax.title.set_text("Sistema 1")
ax.set_xlabel("Eixo x")
plt.show()
```

* Para o seguinte sistema $a + b = 10$ e $2a + 2b = 20$, temos o seguinte plot abaixo. Nele, as duas retas se cruzam, porquê cada ponto é uma solução possível para ambas as equações

```python
import matplotlib.pyplot as plt
import numpy as np

# Resolvendo o sistema linear
a = np.linspace(-0, 10, 100)
b = 10 - a

a2 = (20 - 2 * b) / 2

fig, ax = plt.subplots()
ax.plot(a, b, '*', label='Equação 1')
ax.plot(a2, b, '-', label='Equação 2')
ax.grid(linestyle='--')
ax.axhline(0, color='black', lw=1)
ax.axvline(0, color='black', lw=1)
ax.title.set_text("Sistema 2")
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
ax.legend()
plt.show()
```

* Enquanto que, para esse sistema $a + b = 10, 2a + 2b = 24$ não existe solução, já que as retas não se cruzam.

```python
import matplotlib.pyplot as plt
import numpy as np

# Resolvendo o sistema linear
a = np.linspace(0, 10, 100)
b = 10 - a

a2 = (24 - 2 * b) / 2

fig, ax = plt.subplots()
ax.plot(a, b, '-', label='Equação 1')
ax.plot(a2, b, '-', label='Equação 2')
ax.grid(linestyle='--')
ax.axhline(0, color='black', lw=1)
ax.axvline(0, color='black', lw=1)
ax.title.set_text("Sistema 3")
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
ax.legend()
plt.show()
```

* Dos três sistemas vistos, o primeiro possui uma única solução (8,2), o segundo infinitas soluções e o terceiro nenhuma solução. Ou seja, o primeiro é completo e não singular, o segundo é redundante e singular, enquanto que o terceiro é contraditório e singular. 

## Singularity vs Non Singularity Matrices

* Os sitemas acima 2 e 3 são singulares, e são linhas paralelas. O sistema 2 possui soluções infinitas enquanto o 3 nenhuma solução.
* "As constantes do sistema não importam para determinar o sistema como singular ou não singular" (já que podemos zeralos e todos tem origem no zero)
* Aqui, as matrizes são representadas pelos coeficientes no sistema de equações.

* Matrizes são como sistemas de equações lineares e podem ser singular e não singulares.

**sistema 1:**
<center> a + b = 0</center>
<center> a + 2b = a</center>
<center> </center>
<center> $$\begin{bmatrix} 1 & 1\\ 1 & 2 \end{bmatrix} $$ </center>

**sistema 2:**
<center>a + b = 0</center>
<center>2a + 2b = 0</center>
<center> $$\begin{bmatrix} 1 & 1\\ 2 & 2 \end{bmatrix} $$ </center>

## Linear Dependence and Independence

* A segunda matriz é singular porquê a segunda equação carrega a mesma informação que a anterior, ou seja, trará as mesmas respostas. Ela é somente um multiplo da primeira. Cada elemento da matriz é um multiplo, nesse caso multiplo de 2, do anterior. A segunda linha é **linearmente dependente** da primeira.
* Na primeira matriz, não existe uma constante que ao multiplicar uma das equações teremos a outra, não são linearmente dependentes, ou seja, nenhuma das duas é multipla da outra.

## The Determinant

* Determinante é uma fórmula que determina se o sistema ou a matriz é singular, quando o resultado for zero, ou não singular, quando o resultado for diferente de zero.

<center> $$\begin{bmatrix} 1 & 1 \\ 2 & 2 \end{bmatrix} $$ </center>

* Essa matriz é linearmente dependente porquê a primeira linha $(1, 1) * 2 = (2, 2)$. Enquanto que a primeira matriz $[[1, 1], [1, 2]]$ não possui um valor que multiplique a primeira linha que reseulta na segunda.

* Considere a seguinte matriz singular:
<center> $$\begin{bmatrix} a & b \\ c & d \end{bmatrix} $$ </center>

* a matriz é singular se $(a, b) * k = (c, d)$. Ou seja $ak = c$ e $bk = d$, que é equivalente a $c/d = d/b = k$. Agora, desconsiderando o k, temos que $ad = bc$, ou seja $ad - bc = 0$ e esse é o **determinante** da matriz. Note que estamos fazendo uma multiplicação diagonal entre $ad$ e $cd$ da matriz. 

* Agora, aplicando o determinante das matrizes a seguir
<center> $$\begin{bmatrix} 1 & 1\\ 2 & 2 \end{bmatrix} $$ </center>

Temos que $(1 * 2) - (2 * 1) = 0$, ou seja, essa matriz é singular (!).

<center> $$\begin{bmatrix} 1 & 1\\ 1 & 2 \end{bmatrix} $$ </center>
Enquanto que $(1 * 2) - (1 * 1) = 1$. Essa matriz é não singular.
